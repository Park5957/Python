```python
#네이버 영화순위

import requests  # 웹 페이지 연결 모듈

from bs4 import BeautifulSoup as bs # 읽어온 웹페이지를 분석하기 쉽게 정제 모듈

import pandas  # 데이터 분석 모듈

html = requests.get('https://movie.naver.com/movie/sdb/rank/rmovie.naver?sel=cur&date=20220217')

soup = bs(html.text)  # BeautifulSoup 을 통해 데이터를 분석하기 용이하게 정제한다. 
# print(soup.select)

# <tr>
# <td class="ac"><img alt="01" height="13" src="https://ssl.pstatic.net/imgmovie/2007/img/common/bullet_r_r01.gif" width="14"/></td>
# <td class="title">
# <div class="tit3">
# <a href="/movie/bi/mi/basic.naver?code=191547" title="나일 강의 죽음">나일 강의 죽음</a>
# </div>
# </td>

# len(soup.select('tr > td.title'))  # tr태그가 붙은 부분들을 모두 추출
movies = soup.select('td.title')
# print(movies[0].text)
points = soup.select('td.point')

# mvs = soup.select('tr > td.title > div > a')  # -> 최종 100 추출 완료

movielist = []
rank  = 1

movies = soup.select('td.title')
points = soup.select('td.point')

for each in range(len(soup.select('td.title'))):
  title = movies[each].text.strip()
  point = points[each].text.strip()
  movielist.append([rank, title, point])
  rank += 1
  
movielist
```

```python
import pandas as pd

dm = pd.DataFrame(movielist, columns = ['순위','영화','평점'])
dm
```

```python
dm.to_excel('bugs.xlsx', index=False)
```
