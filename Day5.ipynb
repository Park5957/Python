```python
# Algorithm
# 알고리즘 이란?
# 문제나 과제를 해결하기 위한 처리 절차를 구체적인 순서로 표현한 아이디어 또는 생각
# 알고리즘 자체는 일상생활에서도 많이 이미 사용하고 있다.
# 요리의 레시피, 음악의 악보, 가전제품 사용설명서 등
# 문자, 기호, 일러스트 등으로 나타낸다.

# 알고리즘과 프로그램
# 알고리즘을 프로그래밍 언어로 기술하면 프로그램이 된다.
# 프로그래밍 언어(파이썬, 자바, 씨..)는 컴퓨터에 지시하기 위한 인공 언어이다.

# 프로그램 작성에서의 알고리즘
# 프로그램 작성은 기획, 설계, 프로그래밍, 디버그 로 진행된다.
# 알고리즘은 설계 단계에서 필요하다.

# 좋은 알고리즘 이란?
# 알기 쉽다.(쉬운이해)
# 속도가 빠르다
# 효율적이다(메모리를 작게 사용한다.)
# 재이용이 쉽다(객체지향프로그래밍)

# 왜 알고리즘을 공부해야 하나
# 좋은 프로그램을 만들기 위해(빠르고, 효율적, 범용성)
# 프로그램의 좋고 나쁨을 판단하기 위해
# 프로그램 작성 과정 전체를 효율화하기 위해
# 프로그램 기술을 향상시키기 위해

# 알고리즘의 조건
# 정확한 결과를 얻을 수 있어야 한다.
# (알고리즘은 문제나 과제를 해결하기 위한 절차)
# 반드시 종료되어야 한다.
# (중간 처리나 절차가 잘못 만들어지면 영원히 끝나지 않는다.)

# 알고리즘의 기본형 3가지
# 순차구조 : 처음부터 순서대로 처리하는 절차
# 예) 심부름
#     1. 슈퍼마겟에 간다.
#     2. 라면을 산다.
#     3. 집으로 돌아온다.
#         실행을 원하는 처리를 위로부터 순선대로 작성한다.

# 선택구조 : 조건식으로 판단해 실행할 절차를 전환하는 구조

```

```python
def sum_all(start,end):
  output = 0
  for i in range(start,end+1): # 마지막 수 까지 포함시키기 위함이다.
    output +=i

  return output

print("0에서 100까지의 합 : ", sum_all(0,100))
print("0에서 1000까지의 합 : ", sum_all(0,1000))
print("50에서 100까지의 합 : ", sum_all(50,100))
print("500에서 1000까지의 합 : ", sum_all(500,1000))
```

```python
# 반복문으로 factorial 구하기
def facto(n) :
  output = 1
  for i in range(1,n+1):
   output *= i
  return output

print("3!:", facto(3))
print("3!:", facto(5))
print("3!:", facto(7))
```

```python
#  재귀로 factorial 구하기
def facto(n) :
  if n == 0:
    return 1
  else :
    return n*facto(n-1)

print("3!:", facto(3))
print("3!:", facto(5))
print("3!:", facto(7))
#  재귀함수는 가급접 사용하지 않는편이 좋다.
#  다만 가독성이 좋은편이며ㅡ 코드가 간결해 보이는 장점이 있다.
```

```python
# Fibonacci numbers(사용처: 토끼의 번식 속도, 바이러스의 감염 속도 등등.)

def fibo(n):
  if n == 0:
      return 0
  elif n == 1 or n == 2:
      return 1
  else:
      return fibo(n - 1) + fibo(n - 2)

print("fibonacci(1) : ", fibo(1))
print("fibonacci(2) : ", fibo(2))
print("fibonacci(3) : ", fibo(3))
print("fibonacci(4) : ", fibo(4))
print("fibonacci(5) : ", fibo(5))
```

```python
#  인풋이 몇개의 값이 들어 올지 모를 때
def add_many(*args):
  result = 0

  for arg in args:
    result += arg
  return result

print(add_many(1,2,3))
print(add_many(1,2,3,4,5))
```

```python
def add_many(choice, *args):
  if choice == "add":
    result = 0
    for i in args:
      result += i
  elif choice == "mul":
    result = 1
    for i in args:
      result = result*i
  return result

print(add_many("add",1,2,3,4,5))
print(add_many("mul",1,2,3))
```

```python
#  함수의 결과는 어제나 하나이다.
def add_and_mul(a,b):
  return a+b,a*b
result = add_and_mul(3,4)
print(result) # 튜플로 병합 가능

n,m = add_and_mul(3,4)
print(n,m) #분리 출력 가능
```

```python
# 매개변수 초기값 설정
# 초기값을 설정한 매개 변수를 맨 뒤에 배치해야 한다.

def say_myself(name,old,man=True):       # man=True 가 매개변수 초기값이다.
  print(f"나의 이름은 {name}입니다.")    # print("나의 이름은 %s 입니다." %name)
  print(f"나의 나이는 {old}살 입니다.")  # print("나는 %d살 입니다." %old)
  if man:
    print("남자입니다.")
    print("-"*10)
  else:
    print("여자입니다.")
    print("-"*10)
say_myself("박응용",27)
say_myself("박응용",27,True)
say_myself("이응용",25,False)
```

```python
# 지역변수 local variable 와 전역변수 golbal variable
a = 1           # 함수 내부와 외부에 a 는 다른 a 다.
def vartest(a):
  a = a + 1
  return a      # 이 과정이 없다면 함 수 내부의 a 를 밖으로 가지고 나올 수 없다.

print(vartest(a))
```

```python
a = 1
def vartest():
  global a      # 전역 변수 선언
  a = a + 1     # 다만 사용 빈도가 높지 않음. 전역변수 선언으로 인해 작성된 코드 전체에 영향을 미치기 때문.

vartest()
print(a)
```

```python
# lambda 는 def를 사용해야 할 정도로 복잡하지 않거나 def 를 사용하지 못하는 경우 사용

def add(a,b):
  return a+b
result = add(3,4)
print(result)

# 위 코드를 람다식을 사용해 아래와 같이 간단하게 구현할 수 있다.
add = lambda a,b: a+b
print(add(3,4))
```
