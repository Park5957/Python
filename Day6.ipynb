```python
#  Linear Search 선형 탐색법

#  보관된 데이터를 맨 앞 순서대로 원하는 값을 찾을 때까지 탐색한다.
#  탐색 처리는 반복구조를 사용하여 기술한다.
#  반복 구조에서는 종료 조건을 잊지 말자.

# 첨자(index) 0  1  2  3  4
# 데이터      4  2  3  5  1   찾을 데이터 : 5

arr = [4,2,3,1,2]

for i in range(len(arr)) :
  ii = arr[i]

  if ii == 5 :
    print(f'{i+1} 번째 요소가 일치')
    break
  elif i == len(arr)-1 :
    print(f"찾을 수 없습니다.")
    break

# # 다른방식
arr2 = [4,2,5,1,2]

for i in range(5) :
  
  if arr2[i] == 5 :
    print(f'{i+1} 번째 요소가 일치')
    break
  elif i == 4 :
    print(f"찾을 수 없습니다.")

# 함수를 이용한 방식

def llist(a,b):
  for i in range(len(b)) :
    if a == b[i] :
      return print(f'{i} 번째 요소가 일치')    
  return print(f"찾을 수 없습니다.")
llist(5,[4,2,5,1,2])

# # 다른 예시
def llist2(a,b):
  for i in range(len(b)) :
    if a == b[i] :
      return print(f'{i} 번째 요소가 일치')
  return print(f"찾을 수 없습니다.")
llist2(5,[4,2,0,5,2])
```

```python
# try :
#   # 예외가 발생할 가능성이 있는 코드
# except :
#   # 예외가 발생했을 때 실행할 코드
# else :
#   # 예외가 발생하지 않았을 때 실행할 코드
# finally:
#   # 언제나 실행되는 코드

try:
  num = int(input("정수를 입력하시기 바랍니다. : "))
except:
  print("잘못 입력했습니다.")  
else:
  print(f"당신이 입력한 정수는 {num}입니다.")
finally:
  print("프로그램이 완전히 종료되었습니다.")
# else 는 다른 대부분의 언어에서는 없다.(try 기준 즉, 이말은 try에 else의 구문을 넣어도 된다.)
# 또한 else 사용은 권장하는 방법이 아니다.

# try 구성의 종류
# try + except
# try + except + else
# try + except + finally
# try + except + else + finally
# try + finally
```

```python
def test():
  print('test() 함수의 첫 줄 입니다.')
  try:
    print("try 구문이 실행되었습니다.")
    return
    print("tey 구문의 return 아래입니다.")
  except:
    print("except 구문이 실행되었습니다.")
  else:
    print("else 구문이 실행되었습니다.")
  finally:
    print("fianlly 구문이 실행되었습니다.") # finally 는 어떤 경우일 지라도 실행된다.
  print("test()함수의 마지막 줄입니다.")

test()
```
