```python
#####################
# 정렬 알고리즘 sort 
#####################

# 데이터를 오름차순 또는 내림 차순으로 나열하는 알고리즘
# 검색엔진 엑셍, 워드 이미 많이 사용하고 있다.

# 단순 선택법 - 정영되지 않은 데이터 중 가장 작은 데이터를 선택하여 맨 앞부터 순서대로
#               정렬해 나가는 알고리즘이다.
#               선택정렬. 선택적 단순 선택법

# 0  1  2  3  4   index 번호
# 12 13 11 14 10  배열의 값

# 먼저 "가장 작은 숫자"의 공을 "첫번째 공"과 교환한다.
# 12 13 11 14 10  
# --          --  ## (--) 교환될 값의 표시
# 10 13 11 14 12  
# ** -- --        ## (**) 은 이동 불가 표시
# 10 11 13 14 12
# ** ** --    --
# 10 11 12 14 13
# ** ** ** -- --
# 10 11 12 13 14  ## 정렬 끝

# 단순선택법 알고리즘

# 1. 탐색 범위에서 최소값을 찾는 처리
# 2. 탐색 범위의 맨 앞 요소와 최소값을 교환하는 처리
```

```python
# 잠정적인 최소값의 인덱스를 저장하는 변수 indexMin 을 준비한다.
# 그때 그때 마다의 최소값의 첨자를 저장할 변수 indexMin를 만든다

# arr[indexMin] 와 요소들의 비교처리는 같은 절차의 반복이다. i 는 0에서 4까지 순차적으로
# 변하면서 반복된다. 이 변하는 구조의 첨자는 변수 k로 나타내 보자
# k의 초기값은 1 최대 4까지이다.
```

단순선택법.png

```python
# 잠정 최소값과 비교하는 요소의 첨자를 변수 k로 치완하여 반복 구조로 만들었다.
# 이대로라면 k 에 무한 루프가 발생된다. 반복 구조는 언제나 종료 조건이 필수이다.

#  i 가 0 일때 k는 1 부터 4 까지 변하게 된다. k는 5가 되는 일은 절대 없다.
#  따라서 종료 조건은
#  k < 5
```

```python
arr = [12,13,11,14,10]
a = 0
# **** 가장 작은 숫자를 첫번값과 교환
# 첫번째 값은 고정 다음 값과 가장 작은 값을 교환
for i in range(len(arr)-1):
  a = i                           # 첫번째 최소값 제외용
  for j in range(i+1, len(arr)):  # 제외된 최소값(index번호로 구분하여 제외시킴)
                                  # 이상 부터 새로운 최소값 찾기용
    if arr[j] < arr[a]:           # 새로운 배열의 값이 
      a = j   
      b = arr[i]
      arr[i] = arr[a]
      arr[a] = b
print(arr)
```

```python
##############
#단순 교환법 bubble sort
##############
#  인접한 데이터를 교환하는 처리를 반복한다.

# 0  1  2  3  4  # index 번호
# 5  3  4  1  2  # data 번호

# 5  3  4  1  2 
#          -- -- # (--) 에 해당하는 두 번호는 이미 정렬된 상태
# 5  3  4  1  2 
#       -- --   # (--) 에 해당하는 두 번호는 정렬이 되지 않은 상태
# 5  3  1  4  2 
#    -- --      # (--) 에 해당하는 두 번호는 정렬이 되지 않은 상태
# 5  1  3  4  2 
# -- --         # (--) 에 해당하는 두 번호는 정렬이 되지 않은 상태
# 1  5  3  4  2 # 1은 정렬이 끝난 상태로 확정
#          -- --# (--) 에 해당하는 두 번호는 정렬이 되지 않은 상태
# 1  5  3  2  4 
#       -- --   # (--) 에 해당하는 두 번호는 정렬이 되지 않은 상태
# .(반복)
# 1  2  3  4  5 #끝. 

# 1. 오른쪽 끝부터 순서대로 인접한 공 2개를 오름차순으로 정렬한다.
# 2. 왼쪽 끝 칸부터 순서대로 들어갈 공을 확정시킨다.

# 0  1  2  3  4  # index 번호
# 5  3  4  1  2  # data 번호

# 오른쪽 끝 인접한 공 2개
# 오른쪽 요소 i     arr[i] >
# 왼쪽 요소 i-1     arr[i-1] <

# 정렬이 이미 되어있는 상태인지의 여부를 확인

# "왼쪽 요소가 오른쪽 요소보다 작은가?"
#  arr[i-1] < arr[i] 
```

단순교환법 biubble sort.png

```python
# 여기까지는 순조롭게 비교와 교환처리가 진행 
# 마지막 i-1이 0이 되고 i 가1이 된다면
# 이미 확정된 arr[0] 와 arr[1] 을 비교하게 된다. 이것은 쓸데없는 낭비가 된다.
# 이는 1>0 이라는 조건식에서 발생하는 문제이다.

# arr[1] 을 확정하는 반복처리 부터는 i>0 이 아닌 i>1 로 바꿔야한다.
# 즉 i>0 은 i>(변수) 의 형태로 변형을 해야 한다.

# i > k
```

정렬 - 단순 교환법 최종.png

```python
arr = [5,3,4,1,2]
for i in range(len(arr)-1,0,-1):                # 오른쪽 부터 검사해야 함으로 역순지정
    for j in range(i):                          # 역순의 값들의 비교지정
        if arr[j] > arr[j+1]:                   # 역순시작 값과 왼쪽 값의 비교
            arr[j],arr[j+1]=arr[j+1],arr[j]     # 역순값이 높다면 왼쪽값과 교환

print(arr)
```

```python
# method 매서드
# 클래스가 가지고 있는 함수를 매서드 라고 부른다.
# 매서드는 첫번째 매개변수로 self를 넣어야 한다.

class Student :
  # 생성자 constructor
  def __init__(self,name,korean,math,eng,sci): #self는 필수
    self.name =name
    self.korean =korean
    self.math =math
    self.eng =eng
    self.sci =sci
  def get_sum(self):
    return self.korean+self.math+self.eng+self.sci

  def get_avg(self):
    return self.get_sum() / 4

  def to_string(self):
    return "{}\t\t{}\t\t{}".format(self.name,self.get_sum(),self.get_avg())

# 학생들 리스트
students = [
            Student("슈가",87,98,88,95),
            Student("진",90,99,86,97),
            Student("카리",99,89,86,98)
] 
print("이름","총점","평균",sep="\t\t")
for student in students :
  print(student.to_string())
```

```python
class FourCal:
  def setdata(self,first,second):
    self.first =first
    self.second =second

  def add(self):
    result = self.first+self.second
    return result

  def sub(self):
    result = self.first-self.second
    return result

  def mul(self):
    result = self.first*self.second
    return result

  def div(self):
    result = self.first/self.second
    return result

a =FourCal()
print(type(a))
a.setdata(4,2)
print(a.second)

b =FourCal()
b.setdata(6,13)
print(b.second)

print(a.add()) # 위a.setdata(4,2) 로 a의 값을 지정함으로서
               # a.first/a.second 가 되었기에
               # def add(self) 를 이용해서 a.add 값은 4+2가 된다.
print(f"{a.sub()}:{b.sub()}")
print(f"{a.mul()}:{b.mul()}")
print(f"{a.div()}:{b.div()}")
```

```python
# 생성자 constructor
# 객체가 생성될 때 자동으로 호툴되는 메서드를 의미한다.
# __init__

class FourCal:
  def __init__(self,first,second):  #자동 생성자.
    self.first =first
    self.second =second

  # def setdata(self,first,second):
  #   self.first =first
  #   self.second =second

  def add(self):
    result = self.first+self.second
    return result

  def sub(self):
    result = self.first-self.second
    return result

  def mul(self):
    result = self.first*self.second
    return result

  def div(self):
    result = self.first/self.second
    return result

a =FourCal(4,2)      #a 를 생성할때 값을 입력하여 생성도 가능하다.
print(a.div())
```

```python
# 상속 inheritance
# 어떤 클래스를 만들때 다른 클래스의 기능을 물려받아 사용한다.

class MoreFourCal(FourCal): #()클래스를 상속해서 가지고 온다.
  def pow(self) :
    result = self.first**self.second
    return result

c =MoreFourCal(6,2)
print(c.pow())
```

```python
# 메서드 오버라이딩 method overriding
# a = FourCal(4,0)
# a.div() 의 결과는 ZeroDivisionError                         
class SafeFourCal(FourCal):
  def div(self):        # 부모 클래스의 내용을 수정하여 자식 클래스에 재적용 시키는 것이다.
    if self.second == 0:
      return 0
    else :
      return self.first/self.second

a = SafeFourCal(4,0)
print(a.sub())
```
